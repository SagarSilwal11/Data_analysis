Talking about this data.This is data of Euro to USD currency converter taken in different dates.It has 5 columns
1. Date:This gives the date in which data is taken
2. open:This gives the opening currency conversion rate that day
3. high:This gives the highest currency conversion rate during that day
4. low:This gives the lowest currency conversion rate in that day
5. close:This gives the closing currency conversion rate that day

```python
df=df.reset_index() # here reseting our index back to original dataframe
df
```

```python
new_df=df.copy()
new_df
```

```python
new_df
```

```python
new_df['date']=pd.to_datetime(new_df['date'])
new_df
```

```python
new_df.dtypes
```

```python
new_df['Year'] = new_df['date'].dt.year
new_df['Month'] = new_df['date'].dt.month
new_df['day'] = new_df['date'].dt.day  
new_df['dayname'] = new_df['date'].dt.day_name()
```

Here i further extract year,month,day,dayname from the date column present in new_df.

```python
new_df
```

```python
mean_high_by_month=new_df.groupby('Month')['high'].mean()
mean_high_by_month
```

```python
mean_high_by_year=new_df.groupby('Year')['high'].mean()
mean_high_by_year.plot(kind='line')
plt.grid(True)
```

Here we can say high was after 2007.5 which mean by above char also we seee the high value after 2008 and this is true by this diagram too.

```python
mean_high_by_dayname=new_df.groupby('dayname')['high'].mean()
mean_high_by_dayname.plot(kind='line')
plt.grid(True)
```

```python
mean_high_by_dates=new_df.groupby('day')['high'].mean()
mean_high_by_dates
```
